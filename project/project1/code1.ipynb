## 2. define utils function

```python
# jaccard : compute the jaccard similarity of lsh signature matrix

def jaccard(arr1, arr2):
    """
    Args:
        arr1(ndarray) 
        arr2(ndarray)
    """
    return np.sum(arr1 == arr2) / arr1.size
```

```python
# test jaccard
arr1 = np.array([2,2,1,3,1])
arr2 = np.array([2,4,1,1,1])
jaccard(arr1, arr2)
```

```python
# find the first row which is 1
def find1(ser):
    """
    Args:
        ser(pd.Series)
    Return: the first row in which the element is 1
    """
    for i in range(ser.size):
        if (ser[i] == 1):
            return i+1
#     return ser.index[ser == 1][0] + 1
```

### divide the signature matrix into b bands, r row.

```python
b = 30
bucket = [{} * b]
total_row = signature_matrix.shape[0]
r = total_row // b

# construct hash function
def hashFunction(ser):
    sum_val = 0
    for i in range(ser.size):
        sum_val += i * ser[i]
    return sum_val % 1024
```

```python
for i in range(b): # traverse band and establish buckets
    if i != b - 1:
        ser = signature_matrix.loc[i * r: i * r - 1].apply(hashFunction)
        for index,item in enumerate(ser):
            bucket[i][item] = index + 1 
    else:
        ser = signature_matrix.loc[i * r: total_row - 1].apply(hashFunction)
        for index, item in enumerate(ser):
            bucket[b - 1][item] = index + 1
```

### Searching

```python
author = int(input("please input the author u want to search:"))
result = []
for i in range(b):
    if i != b - 1:
        ser = signature_matrix.loc[i * r: (i + 1) * r - 1, author]
        val = hashFunction(ser)
        result += set(bucket[i][val].values()) - author
    else:
        ser = signature_matrix.loc[i * r: total_row, author]
        val = hashFunction(ser)
        result += set(bucket[i][val].values()) - author
print(f"candidates are {result}")
```
