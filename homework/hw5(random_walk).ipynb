1. 给定一个转移矩阵P和状态向量$\pi$
$$
P = \begin{pmatrix} \frac{1}{2} \frac{1}{2} \\ \frac{1}{4} \frac{3}{4} \end{pmatrix},
\pi = \begin{pmatrix} 1 \\ 0 \end{pmatrix}^T
$$

(1) 计算$\pi P, \pi P^2, \pi P^3 和 \pi P^4$。

```python
import numpy as np

p = np.array([[1/2, 1/2], [1/4, 3/4]])
pi = np.array([1,0])

print(f"pi * P = {pi @ p}")
print(f"pi * P^2 = {pi @ p @ p}")
print(f"pi * P^3 = {pi @ p @ p @ p}")
print(f"pi * P^4 = {pi @ p @ p @ p @ p}")
```

(2)证明$\pi P^n$的结果接近一个常数向量。

解： 因为状态转移矩阵P对应的马氏链是不可约的，且是反周期的，所以根据书中的定理7.2
$\lim\limits_{n \to \infty}P_{ij}^n = \pi(j) \$

因此$\pi P^n$的结果接近一个常数向量。

实验：

```python
N = 100
for i in range(1, N + 1):
    res = pi
    for j in range(i):
        res = res @ p 
    if (i > 95):
        print(f"pi * P^{i} = {res}")
```
